#### locate specific
##### loc()
##### iloc() -> almost same as loc() but the upper index is not inclusive, where with loc() it was inclusive

```python
# coffee.loc[#Rows , #Columns]
coffee.loc[0]
```

```python
coffee.loc[[0,1,2]]
```

```python
#slicing
print(coffee.loc[0:5])
print(coffee.loc[0:])
print(coffee.loc[5:])
print(coffee.loc[::2])
```

```python
print(coffee.loc[0:5,["Day","Units Sold"]])
# upper index is inclusive
```

```python
print(coffee.iloc[0:5, [0, 2]]) # used to access rows and columns but only using index value
# upper index is NOT inclusive
```

```python
# coffee.index is already set to 'Day' (as shown in your variable summary)
# To access all rows where the index is 'Day', use .loc with the actual day name, e.g. "Monday"
# If you want to see all rows for "Monday":
coffee.loc["Monday":"Friday","Units Sold"]
```

### Sorting of data

```python
# To sort by multiple columns, pass a list to 'by'
# coffee.sort_values(by=["Units Sold", "Coffee Type"], ascending=False)
coffee.sort_values(by=["Units Sold", "Coffee Type"], ascending=[0,1]) # 0 for units sold for not ascending and 1 for coffee type for ascending
```

### Iterating over a dataframe (df) with For Loop (df.iterrows())

```python
for index,row in coffee.iterrows():
    print(index)
    print(row)
    print("\n\n")
```

```python
for index,row in coffee.iterrows():
    print(index)
    print(row["Units Sold"])
    print("\n\n")
```
